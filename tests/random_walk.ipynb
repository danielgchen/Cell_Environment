```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
n_steps = int(1e3)
resolutions = [3 ** exp for exp in range(1, 7)]
colors = sns.color_palette('tab10').as_hex()
# > create plot for resolutions
fig,ax = plt.subplots(figsize=[12,20])
for idc, resolution in enumerate(resolutions):
    rng = np.random.default_rng(0)
    steps = np.linspace(-1, 1, resolution)
    xs = [rng.choice(steps) for _ in range(n_steps)]
    ys = [rng.choice(steps) for _ in range(n_steps)]
    x,y = 0,0  # track current position
    xs_plot,ys_plot = [x],[y]
    ax.scatter(x, y, color=colors[idc])
    for idx in range(n_steps):
        x += xs[idx]
        y += ys[idx]
        ax.scatter(x, y, color=colors[idc])
        xs_plot.append(x)
        ys_plot.append(y)
    ax.plot(xs_plot, ys_plot, color=colors[idc], label=resolution)
# > get uniform
rng = np.random.default_rng(0)
xs = [rng.uniform(-1, 1) for _ in range(n_steps)]
ys = [rng.uniform(-1, 1) for _ in range(n_steps)]
x,y = 0,0  # track current position
xs_plot,ys_plot = [x],[y]
ax.scatter(x, y, color='k')
for idx in range(n_steps):
    x += xs[idx]
    y += ys[idx]
    ax.scatter(x, y, color='k')
    xs_plot.append(x)
    ys_plot.append(y)
ax.plot(xs_plot, ys_plot, color='k', label='uniform')
ax.legend()
```
