```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Define a brute force method for causing increasd turbulence close to the vent

```python
resolution = 4
n_dimensions = 2
window_width = 10
n_points = 100
rng = np.random.default_rng(0)
values_list = np.zeros((1,n_dimensions))
for _ in range(n_points):
    diffs = np.array([rng.integers(-resolution, resolution) / resolution for _ in range(n_dimensions)])
    new_points = values_list[-1] + (diffs * (window_width / (np.linalg.norm(values_list[-1]) + 1)))
    values_list = np.append(values_list, [new_points], axis=0)
```

```python
fig,ax = plt.subplots(figsize=[5,10])
ax.grid(False)
ax.scatter(values_list[:,0], values_list[:,1])
ax.plot(values_list[:,0], values_list[:,1])
```
